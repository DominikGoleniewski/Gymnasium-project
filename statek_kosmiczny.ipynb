```python
import gymnasium as gym
import numpy as np

from stable_baselines3 import DQN
```

```python
model = DQN(
    "MlpPolicy",
    "LunarLander-v2",
    verbose=1,
    exploration_final_eps=0.1,
    target_update_interval=250,
)
```

```python
from stable_baselines3.common.evaluation import evaluate_policy
```

Pokazanie działania niewytrenowanego agenta.


```python

eval_env = gym.make("LunarLander-v2", render_mode="rgb_array")


mean_reward, std_reward = evaluate_policy(
    model,
    eval_env,
    n_eval_episodes=10,
    deterministic=True,
)

print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

```python
import imageio
import numpy as np

from stable_baselines3 import A2C


images = []
obs, info = eval_env.reset()
img = eval_env.render()
for i in range(1000):
    images.append(img)
    action, _ = model.predict(obs)
    obs, _, _ ,_,_ = eval_env.step(action)
    img = eval_env.render()

imageio.mimsave("nie_wytrenowany.gif", [np.array(img) for i, img in enumerate(images) if i%2 == 0], fps=29)
```

## Trenowanie modelu oraz zapisanie wytrenowanego modelu


```python
model.learn(total_timesteps=int(1e5))
model.save("dqn_lunar")
del model
```
